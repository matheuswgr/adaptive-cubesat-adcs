```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv("adaptive_controller.csv")

plt.show()

fig,ax = plt.subplots(figsize=(20,10))

ax.set_title("Filtered Attitude")
ax.plot(df[' filtered attitudew'])
ax.plot(df['filtered attitudex'])
ax.plot(df['filtered attitudey'])
ax.plot(df['filtered attitudez'])

ax.legend(["w","x","y","z"])
plt.show()
fig,ax = plt.subplots(figsize=(20,10))

ax.set_title("Attitude Error")
ax.plot(df['error attitudew'])
ax.plot(df['error attitudex'])
ax.plot(df['error attitudey'])
ax.plot(df['error attitudez'])

ax.legend(["w","x","y","z"])
plt.show()


fig,ax = plt.subplots(figsize=(20,10))

ax.set_title("Sliding Variable")
ax.plot(df[" sliding variablex"])
ax.plot(df["sliding variabley"])
ax.plot(df[" sliding variablez"])

ax.legend(["x","y","z"])
plt.show()

fig,ax = plt.subplots(figsize=(20,10))

ax.set_title("Control Signal")
ax.plot(df[" control signalx"])
ax.plot(df["control signaly"])
ax.plot(df["control signalz "])

ax.legend(["x","y","z"])
plt.show()

fig,ax = plt.subplots(figsize=(20,10))

ax.set_title("Filtered Frame Velocity")
ax.plot(df[" filtered velocityx"])
ax.plot(df["filtered velocityy"])
ax.plot(df["filtered velocityz"])
```
